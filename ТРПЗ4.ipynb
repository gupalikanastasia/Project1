1. Що таке об'єктно-орієнтоване програмування (ООП)?

ООП — це парадигма (стиль) програмування, яка будується навколо поняття "об'єктів", що поєднують у собі дані (атрибути) та методи (функції) для роботи з цими даними. Це спосіб моделювання реального світу в коді.

2. Які основні принципи ООП?

Існує чотири "кити" ООП:

* Інкапсуляція: приховування внутрішньої реалізації та даних від прямого доступу ззовні.

* Спадкування: можливість створювати нові класи на основі існуючих.

* Поліморфізм: здатність об'єктів різних класів виконувати дію з однаковим ім'ям, але по-різному.

* Абстракція: виділення головних характеристик об'єкта та ігнорування менш важливих деталей.

3. Що таке клас та об'єкт?

* Клас: це шаблон, креслення або схема (наприклад, креслення автомобіля). Він описує, які властивості та поведінку матимуть об'єкти.

* Об'єкт: це конкретний екземпляр класу (наприклад, конкретний червоний BMW у гаражі), створений за цим шаблоном.

4. Що таке атрибути та методи?

* Атрибути: це змінні, що належать об'єкту (характеристики). Наприклад: color, model, age.

* Методи: це функції, що належать об'єкту (дії). Наприклад: drive(), stop(), speak().

5. Що таке конструктор?

Конструктор — це спеціальний метод, який автоматично викликається при створенні нового об'єкта. В Python це метод __init__. Він використовується для початкового налаштування об'єкта (ініціалізації атрибутів).

6. Що таке інкапсуляція та як вона реалізується в Python?

Інкапсуляція — це механізм, що об'єднує дані та код, який ними маніпулює, і захищає їх від зовнішнього втручання. В Python:

* Public (публічні): доступні всюди (назва без підкреслень).

* Protected (захищені): _name (одне підкреслення) — умовна домовленість, що це для внутрішнього користування.

* Private (приватні): __name (два підкреслення) — інтерпретатор змінює ім'я змінної, ускладнюючи прямий доступ ззовні.

7. Що таке спадкування та як воно реалізується в Python?

Це процес, коли один клас (нащадок) переймає атрибути та методи іншого класу (батька). Синтаксис: class Student(Person): — тут Student успадковує все від Person.

8. Що таке поліморфізм та як він реалізується в Python?

Це здатність різних об'єктів реагувати на один і той самий виклик методу по-своєму. В Python: Якщо класи Cat і Dog мають метод speak(), ми можемо викликати animal.speak() для будь-якого з них, не знаючи заздалегідь, хто це саме, і кожен видасть свій звук.

9. Що таке статичні методи та атрибути?

* Статичні атрибути: належать самому класу, а не окремим об'єктам. Вони спільні для всіх екземплярів (наприклад, лічильник створених об'єктів).

* Статичні методи: методи, яким не потрібен доступ до конкретного об'єкта (self) або класу (cls). В Python позначаються декоратором @staticmethod.

10. Що таке абстрактні класи та методи?

* Абстрактний клас: клас, з якого не можна створити об'єкт. Він слугує лише базою для інших класів.

* Абстрактний метод: метод, який оголошений в абстрактному класі, але не має реалізації. Класи-нащадки зобов'язані його реалізувати. В Python для цього використовується модуль abc.

11. Які переваги використання ООП?

* Структурованість коду (легше читати та підтримувати великі проекти).

* Повторне використання коду (через спадкування).

* Безпека даних (через інкапсуляцію).

12. Які недоліки використання ООП?

* Більший розмір програми (більше коду).

* Може бути повільнішим за процедурне програмування.

* Вищий поріг входження (складніше для новачків зрозуміти архітектуру).

```python
class Person:
    total_count = 0

    def __init__(self, name, age):
        self.__name = name
        self.__age = age
        Person.total_count += 1

    def get_name(self):
        return self.__name

    def set_name(self, new_name):
        self.__name = new_name

    def get_age(self):
        return self.__age

    def set_age(self, new_age):
        if new_age > 0:
            self.__age = new_age
        else:
            print("Вік має бути додатнім числом!")

    def introduce(self):
        print(f"Привіт! Мене звати {self.__name}, мені {self.__age} років.")

    @staticmethod
    def get_total_count():
        return Person.total_count


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def introduce(self):
        print(f"Привіт! Я студент. Мене звати {self.get_name()}, мій ID: {self.student_id}.")


class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def introduce(self):
        print(f"Доброго дня! Я викладач {self.get_name()}. Я викладаю {self.subject}.")


print("Створення об'єктів та Інкапсуляція")
person1 = Person("Олег", 30)
person1.introduce()
person1.set_name("Олександр")
print(f"Оновлене ім'я: {person1.get_name()}, Вік: {person1.get_age()}")

print("\nСпадкування та Створення Студента")
student1 = Student("Марія", 19, "S12345")
student1.introduce()

print("\nПоліморфізм")
teacher1 = Teacher("Віктор Петрович", 45, "Фізика")
person2 = Person("Ірина", 25)

people_list = [person1, student1, teacher1, person2]

for human in people_list:
    human.introduce()

print("\nСтатичні методи та атрибути")
print(f"Всього створено об'єктів Person (включаючи студентів та вчителів): {Person.get_total_count()}")
```
