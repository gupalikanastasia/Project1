1. П'ять принципів SOLID:

S — Single Responsibility Principle (Принцип єдиної відповідальності): Клас повинен мати лише одну причину для змін. Це означає, що клас має виконувати лише одну задачу або мати одну зону відповідальності.

O — Open/Closed Principle (Принцип відкритості/закритості): Програмні сутності (класи, модулі, функції) повинні бути відкриті для розширення, але закриті для змін.

L — Liskov Substitution Principle (Принцип підстановки Лісков): Об'єкти в програмі повинні бути замінюваними екземплярами їхніх підтипів без зміни правильності виконання програми.

I — Interface Segregation Principle (Принцип розділення інтерфейсу): Клієнти не повинні залежати від інтерфейсів, які вони не використовують. Краще мати багато вузькоспеціалізованих інтерфейсів, ніж один загальний.

D — Dependency Inversion Principle (Принцип інверсії залежностей): Модулі вищого рівня не повинні залежати від модулів нижчого рівня. Обидва повинні залежати від абстракцій.

2. Який принцип закликає до створення класів з єдиною відповідальністю? Це Single Responsibility Principle (SRP). Він означає, що всі методи класу мають бути сфокусовані на виконання одного спільного завдання. Якщо є методи, які не відповідають меті існування класу, їх треба винести за його межі.

```python
class Employee:
    def __init__(self, name, position, salary):
        self.name = name
        self.position = position
        self.salary = salary

    def save_employee(self):
        print(f"Співробітника {self.name} збережено в базу даних.")

    def generate_report(self):
        print(f"ЗВІТ ПРО СПІВРОБІТНИКА:\nІм'я: {self.name}\nПосада: {self.position}\nЗарплата: ${self.salary}")

emp = Employee("Олег", "Розробник", 3000)
emp.save_employee()
emp.generate_report()
```

```python
class Employee:
    def __init__(self, name, position, salary):
        self.name = name
        self.position = position
        self.salary = salary

    def save(self):
        print(f"Дані {self.name} збережено.")

class ReportGenerator:
    def generate(self, employee: Employee):
        print(f"ЗВІТ")
        print(f"Співробітник: {employee.name}")
        print(f"Позиція: {employee.position}")
        print(f"Оклад: {employee.salary}")

emp = Employee("Марія", "Менеджер", 2500)
report_gen = ReportGenerator()

emp.save()
report_gen.generate(emp)
```

```python
import math

class AreaCalculator:
    def calculate_area(self, shape):
        if shape.type == "circle":
            return math.pi * (shape.radius ** 2)
        elif shape.type == "rectangle":
            return shape.width * shape.height

class CircleStruct:
    def __init__(self, radius):
        self.type = "circle"
        self.radius = radius

class RectStruct:
    def __init__(self, width, height):
        self.type = "rectangle"
        self.width = width
        self.height = height
```

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height

class AreaCalculator:
    def calculate(self, shape: Shape):
        return shape.calculate_area()

circle = Circle(5)
rect = Rectangle(4, 6)

calculator = AreaCalculator()

print(f"Площа кола: {calculator.calculate(circle):.2f}")
print(f"Площа прямокутника: {calculator.calculate(rect)}")
```

## Відповіді на запитання
1. Які проблеми виникають при порушенні принципів SOLID?

Жорсткість (Rigidity): Будь-яка дрібна зміна викликає каскад змін в інших частинах програми.

Крихкість (Fragility): Зміни в одному місці несподівано ламають функціонал в іншому, здавалося б, не пов'язаному місці.

Неможливість повторного використання: Код настільки сильно пов'язаний (high coupling), що його неможливо використати в іншому проекті без тягнення за собою "зайвого багажу".

2. Як принципи SOLID допомагають покращити якість коду та його супровід? Вони роблять код модульним. Кожен клас стає маленьким, зрозумілим "цеглинкою". Це спрощує:

Тестування: Легше написати тест для класу, який робить одну річ (SRP).

Масштабування: Легко додавати нові фічі без переписування ядра (OCP).

Читабельність: Новим розробникам легше зрозуміти систему частинами.

3. Які є найпоширеніші помилки при застосуванні принципів SOLID?

Over-engineering (Надмірне ускладнення): Спроба застосувати всі принципи відразу там, де це не потрібно (наприклад, у дуже простих скриптах). Створюється купа інтерфейсів та класів заради однієї простої функції.

Дроблення логіки: Доведення SRP до абсурду, коли кожен метод виноситься в окремий клас, через що код стає важко відстежувати ("спагетті з класів").

Неправильне визначення абстракцій: Створення абстракцій, які не відповідають реальній суті об'єктів (порушення LSP).
